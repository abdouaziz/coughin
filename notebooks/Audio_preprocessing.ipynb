```python
import pandas as pd
import librosa , librosa.display
import matplotlib.pyplot as plt
from IPython import  display
import numpy as np
```

```python
file = "b.mp3"
```

```python
 display.Audio(file)
```

```python
signal , sr  = librosa.load(file , sr=22050) #The signal could  22050 * 3 coefficients
```

```python
librosa.display.waveplot(signal)
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.show()
```

```python
fft = np.fft.fft(signal)
magnitude = np.abs(fft)
frequency = np.linspace(0 , sr , len(magnitude))
plt.plot(frequency , magnitude)
plt.xlabel("Frequency")
plt.ylabel("Magnitude")
plt.show()
```

```python
#Here we can have the half of spectrum
```

```python
left_magnitude = magnitude[:int(len(magnitude)/2)]
left_frequency = frequency[:int(len(frequency)/2)]
plt.plot(left_frequency , left_magnitude)
plt.xlabel("Frequency")
plt.ylabel("Magnitude")
plt.show()
```

```python
#Here we going for the spectogram
```

```python
n_fft = 2048
hop_length = 512 

stft =  librosa.core.stft(signal , n_fft=n_fft , hop_length=hop_length )
spectogram = np.abs(stft)
librosa.display.specshow(spectogram , sr=sr , hop_length=hop_length)
plt.xlabel("Time")
plt.ylabel("Frequency")
plt.colorbar()
plt.show()
```

```python
#Here with the log spectogram
log_spectogram = librosa.amplitude_to_db(spectogram)
librosa.display.specshow(log_spectogram , sr=sr , hop_length=hop_length)
plt.xlabel("Time")
plt.ylabel("Frequency")
plt.colorbar()
plt.show()
```

```python
#Here for MFFCs
```

```python
MFFCs = librosa.feature.mfcc(signal , n_fft = n_fft , hop_length=hop_length ,  n_mfcc=13)
librosa.display.specshow(MFFCs , sr= sr , hop_length=hop_length)
plt.xlabel("Time")
plt.ylabel("Frequency")
plt.colorbar()
plt.show()
```
